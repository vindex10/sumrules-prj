```python
%numerics
import sumrules.lib.legacy_analytics as alyt
import sumrules.lib.evaluators as evs
from sumrules.utils import parallel
import sumrules
```

```python
sumrules.constants["eps"] = 0.001
```

# Discrete 

```python
mpd = evs.McolPDiscEvaluator(None, None, alyt.energColDisc)

mpd.vectorized = True
mpd.mapper = parallel.npMap
```

## MP2

```python
mpd.MP = alyt.sqedMP2
mpd.psiColP = alyt.psiColPdisc2
```

```python
def make(plt, par):
    mpd.CONST["m"], mpd.CONST["g"] = par
    points = sp.exp(sp.linspace(sp.log(1e-2), sp.log(100), 250))

    res = sp.empty_like(points)
    for i in range(len(points)):
        mpd.params({"maxP": points[i]})
        res[i] = sp.absolute(mpd.compute(3,2))
        
    return plt.plot(points, res, ".", label="m={0}, g={1}".format(*par))

plt.xscale("log")
p1, = make(plt, (1.42, -0.6)) # charm
p2, = make(plt, (1.27, -0.6)) # central
p3, = make(plt, (4.2, -0.6)) # bottom
plt.legend(handles=[p1, p2, p3])
plt.figtext(0.02, 0.02, "Computed for several m and constant g")
```

# Continuous 

```python
mp = evs.McolPEvaluator(None, alyt.psiColP)

mp.vectorized = True
mp.mapper = parallel.npMap
mp.absErr = 1e-8
```

```python
mpw = evs.McolPEvaluator(None, lambda k,p,Tkp: (2*sp.pi)**3*sumrules.constants["eps"]/(sp.pi)**2/(sumrules.constants["eps"]**2 + p**2 + k**2 - 2*p*k*sp.cos(Tkp))**2)
mpw.vectorized = True
mpw.mapper = parallel.npMap
mpw.absErr = 1e-8
```

## MP2

```python
mp.MP = alyt.sqedMP2
mpw.MP = alyt.sqedMP2
```

```python
def make(ax, par, p):
    axTot, axPw = ax
    
    mp.CONST["m"], mp.CONST["g"] = par
    points = sp.exp(sp.linspace(sp.log(1e-2), sp.log(1e3), 100))

    res = sp.empty_like(points, dtype=sp.complex128)
    resPw = sp.empty_like(points, dtype=sp.complex128)
    for i in range(len(points)):
        mp.params({"maxP": points[i]})
        mpw.params({"maxP": points[i]})
        res[i] = mp.compute(p, sp.sqrt(p**2 + mp.CONST["m"]**2), 0.2, 0)
        resPw[i] = mpw.compute(p, sp.sqrt(p**2 + mp.CONST["m"]**2), 0.2, 0)
    
    axTot.plot(points, sp.absolute(res), ".", label="m={0}, g={1}, p={2}".format(*par, p))
    axPw.plot(points, sp.absolute(res-resPw), ".", label="m={0}, g={1}, p={2}".format(*par, p))
```

```python
fig, ax = plt.subplots(1, 2)
fig.set_size_inches(18, 8)

make(ax, (1.42, -0.6), 2.5) # charm
make(ax, (1.27, -0.6), 2.5) # central
make(ax, (4.2, -0.6), 2.5) # bottom

for a in ax:
    a.set_xscale("log")
    a.legend()

fig.text(0.02, 0.02, "Computed for several m and constant g")
plt.show()
```
